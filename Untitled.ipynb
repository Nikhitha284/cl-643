```python
def write_gams_data(states, demands, units, tasks, task_data, changeover_data):
    gams_code = ""
    
    # Create a mapping of unit to tasks
    unit_to_tasks = {}
    for task in task_data:
        task_name, unit, state, min_amount, min_time = task
        if unit not in unit_to_tasks:
            unit_to_tasks[unit] = []
        unit_to_tasks[unit].append(task_name)

    # Define sets
    gams_code += f"Sets\n    s /{','.join(states)}/\n"
    gams_code += f"    u /{','.join(units)}/\n"
    gams_code += f"    t /{','.join(tasks)}/;\n\n"

    # Define demands (inject demands per state)
    gams_code += "Parameters\n    Demand(s) /\n"
    for i, state in enumerate(states):
        gams_code += f"    {state} {demands[i]}\n"
    gams_code += "    /;\n\n"

    # Task characteristics: Task_Unit_State_MinAmount_MinTime
    gams_code += "Parameters\n"
    gams_code += "    MinAmount(t), MinTime(t), ProductionRate(t);\n\n"

    # Collect SuitableUnit and ProducedState values
    suitable_unit_values = []
    produced_state_values = []

    for task in task_data:
        task_name, unit, state, min_amount, min_time = task
        production_rate = float(min_amount) / float(min_time)

        # Generate GAMS code for MinAmount, MinTime, and ProductionRate
        gams_code += f"MinAmount('{task_name}') = {min_amount};\n"
        gams_code += f"MinTime('{task_name}') = {min_time};\n"
        gams_code += f"ProductionRate('{task_name}') = {production_rate};\n\n"

        # Collect SuitableUnit and ProducedState for binary assignments
        suitable_unit_values.append(f"    SuitableUnit('{task_name}', '{unit}') = 1;")
        produced_state_values.append(f"    ProducedState('{task_name}', '{state}') = 1;")

    # Add SuitableUnit values in binary format
    gams_code += "Parameter SuitableUnit(t, u);\n"
    gams_code += "\n".join(suitable_unit_values) + "\n\n"

    # Add ProducedState values in binary format
    gams_code += "Parameter ProducedState(t, s);\n"
    gams_code += "\n".join(produced_state_values) + "\n\n"

    gams_code += "Parameters\n    aij(t, t, u), a0i(t, u), Tmax(t);\n\n"

    for unit in units:
        if unit in unit_to_tasks:
            for task in unit_to_tasks[unit]:
                # Retrieve the state directly from the task data
                state = next((t[2] for t in task_data if t[0] == task), None)
                if state is not None:
                    demand_value = demands[states.index(state)]  # Get demand for the corresponding state
                    production_rate = f"ProductionRate('{task}')"
                    gams_code += f"Tmax('{task}') = {demand_value} / {production_rate};\n"

            # Generating changeover times for different tasks within the unit
            for task1 in unit_to_tasks[unit]:
                for task2 in unit_to_tasks[unit]:
                    for changeover in changeover_data:
                        if (changeover[0] == task1 and changeover[1] == task2):
                            gams_code += f"aij('{task1}', '{task2}', '{unit}') = {changeover[2]};\n"
            for task1 in unit_to_tasks[unit]:
                gams_code += f"a0i('{task1}', '{unit}') = aij('{task1}', '{task1}', '{unit}');\n"

    gams_code += ";\n"
    return gams_code


gams_code_with_data = write_gams_data(states, demands, units, tasks, task_data, changeover_data)
print(gams_code_with_data)
```

```python
# Specify the file path
file_path = 'output.txt'

# Write data to the file
with open(file_path, 'w') as file:
    file.writelines(gams_code_with_data)

print(f"Data written to {file_path}")
```
